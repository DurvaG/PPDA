```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 
%matplotlib inline 

## Models
from sklearn import linear_model
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import Perceptron
from sklearn.linear_model import SGDClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC, LinearSVC
from sklearn.naive_bayes import GaussianNB
from xgboost import XGBClassifier
import xgboost as xgb
from sklearn.ensemble import AdaBoostClassifier
from catboost import CatBoostClassifier
import lightgbm as lgb 
from keras.models import Sequential
from keras.layers import Dense
from sklearn.neural_network import MLPClassifier

## Model evaluators
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.model_selection import RandomizedSearchCV, GridSearchCV
from sklearn.metrics import confusion_matrix, classification_report
from sklearn.metrics import precision_score, recall_score, f1_score
from sklearn.metrics import plot_roc_curve
```

```python
train_df = pd.read_csv("TrainingData.csv")
train_df.head(10)
```

```python
train_df1 = train_df.drop(['mvar47'], axis=1)

train_df2 = train_df1.replace(to_replace ="[a-zA-Z]+", value = np.nan, regex = True)
train_df2 = train_df2.astype('float')
train_df2 = pd.concat([train_df2, train_df['mvar47']], axis=1)
train_df2.info()
```

```python
train_df2.head(10)
```

### One hot encoding to convert categorical feature to numerical one

```python
train_df3 = pd.get_dummies(train_df2, columns = ['mvar47'])
train_df3.head()
```

```python
(train_df2.isnull().sum()/train_df2.shape[0] * 100.00).round(2).sort_values(ascending=False)
```

```python
# dropping columns which have > 70% NULL values
train_df4 = train_df3.drop(['mvar31', 'mvar40'], axis=1)
```

```python
from sklearn.impute import KNNImputer

imputer = KNNImputer(n_neighbors=3)
imputed = imputer.fit_transform(train_df4)
train_data = pd.DataFrame(imputed, columns=train_df4.columns)
```

```python
X = train_data.drop(['default_ind'], axis = 1)
y = train_data['default_ind']

from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
from sklearn.metrics import auc, average_precision_score
from sklearn.metrics import precision_recall_curve
```

```python
LightGBM = lgb.LGBMClassifier()
LightGBM.fit(X_train, Y_train)
Y_pred = LightGBM.predict(X_test)
precision, recall, _ = precision_recall_curve(y_test, Y_pred)
score_lgbm = auc(recall, precision)

sgd = linear_model.SGDClassifier(max_iter=5, tol=None)
sgd.fit(X_train, Y_train)
Y_pred = sgd.predict(X_test)
precision, recall, _ = precision_recall_curve(y_test, Y_pred)
score_sgd = auc(recall, precision)

logreg = LogisticRegression(max_iter=10000)
logreg.fit(X_train, Y_train)
Y_pred = logreg.predict(X_test)
precision, recall, _ = precision_recall_curve(y_test, Y_pred)
score_log = auc(recall, precision)

knn = KNeighborsClassifier(n_neighbors = 3) 
knn.fit(X_train, Y_train)  
Y_pred = knn.predict(X_test)  
precision, recall, _ = precision_recall_curve(y_test, Y_pred)
score_knn = auc(recall, precision)

gaussian = GaussianNB() 
gaussian.fit(X_train, Y_train)  
Y_pred = gaussian.predict(X_test)
precision, recall, _ = precision_recall_curve(y_test, Y_pred)
score_gaussian = auc(recall, precision)

perceptron = Perceptron(max_iter=100)
perceptron.fit(X_train, Y_train)
Y_pred = perceptron.predict(X_test)
precision, recall, _ = precision_recall_curve(y_test, Y_pred)
score_perceptron = auc(recall, precision)

decision_tree = DecisionTreeClassifier() 
decision_tree.fit(X_train, Y_train)  
Y_pred = decision_tree.predict(X_test)  
precision, recall, _ = precision_recall_curve(y_test, Y_pred)
score_decision_tree = auc(recall, precision)

random_forest = RandomForestClassifier(n_estimators=100)
random_forest.fit(X_train, Y_train)
Y_pred = random_forest.predict(X_test)
precision, recall, _ = precision_recall_curve(y_test, Y_pred)
score_random_forest = auc(recall, precision)

AdaBoost = AdaBoostClassifier()
AdaBoost.fit(X_train, Y_train)
Y_pred = AdaBoost.predict(X_test)
precision, recall, _ = precision_recall_curve(y_test, Y_pred)
score_AdaBoost = auc(recall, precision)

XGBoost = XGBClassifier()
XGBoost.fit(X_train, Y_train)
Y_pred = XGBoost.predict(X_test)
precision, recall, _ = precision_recall_curve(y_test, Y_pred)
score_XGBoost = auc(recall, precision)

CatBoost = CatBoostClassifier()
CatBoost.fit(X_train, Y_train)
Y_pred = CatBoost.predict(X_test)
precision, recall, _ = precision_recall_curve(y_test, Y_pred)
score_CatBoost = auc(recall, precision)

from sklearn.neural_network import MLPClassifier

MLPClassifier = MLPClassifier()
MLPClassifier.fit(X_train, Y_train)
Y_pred = MLPClassifier.predict(X_test)
precision, recall, _ = precision_recall_curve(y_test, Y_pred)
score_MLPClassifier = auc(recall, precision)

# comparing all the models

results = pd.DataFrame({
    'Model': ['KNN', 'Logistic Regression','Random Forest', 'Naive Bayes', 'Perceptron', 
              'Stochastic Gradient Decent','Decision Tree', 'XG Boost', 'AdaBoost', 'Neural Network', 'Cat Boost', 'Light GBM'],
    'PR-AUC Score': [score_knn, score_log, score_random_forest, score_gaussian, score_perceptron, score_sgd, 
              score_decision_tree, score_XGBoost, score_AdaBoost, score_MLPClassifier, score_CatBoost, score_lgbm]})
result_df = results.sort_values(by='PR-AUC Score', ascending=False)
result_df = result_df.set_index('PR-AUC Score')
result_df
```
